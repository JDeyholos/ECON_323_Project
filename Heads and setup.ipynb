```python
# Put years at the top for GDP, then align both datasets to the same years

# 1) Internet dataset already has year columns
internet_years = sorted([c for c in df.columns if str(c).isdigit()])

# 2) GDP dataset has years in rows; pivot one GDP variable into year columns
# Change 'rgdpe' to another GDP variable if needed.
gdp_wide = gdp.pivot_table(index='country', columns='year', values='rgdpe', aggfunc='first').reset_index()
gdp_wide.columns = [str(c) for c in gdp_wide.columns]  # make year headers strings

gdp_years = sorted([c for c in gdp_wide.columns if c.isdigit()])
common_years = sorted(set(internet_years).intersection(gdp_years))

print('Internet years:', internet_years[0], 'to', internet_years[-1], '| n =', len(internet_years))
print('GDP years:', gdp_years[0], 'to', gdp_years[-1], '| n =', len(gdp_years))
print('Common years:', common_years[0], 'to', common_years[-1], '| n =', len(common_years))

# 3) Keep only matching years so both tables line up
df_aligned = df[['Country Name'] + common_years].copy()
gdp_aligned = gdp_wide[['country'] + common_years].copy()

# Optional: merge after alignment
merged_df = pd.merge(df_aligned, gdp_aligned, left_on='Country Name', right_on='country', how='inner')

print('\nInternet aligned preview:')
display(df_aligned.head())
print('\nGDP aligned preview:')
display(gdp_aligned.head())
print('\nMerged aligned preview:')
display(merged_df.head())
```
